```python
def get_outliers(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1

    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr

    return column[(column < lower) | (column > upper)].index
```

```python
outliers_index = get_outliers(Mov['Timing'])
df_clean = Mov.copy()
df_clean['Timing'].loc[outliers_index] = Mov['Timing'].median()
```
